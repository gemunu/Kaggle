```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor, BaggingRegressor
import snowballstemmer
```

```python
df_train = pd.read_csv('train.csv', encoding="ISO-8859-1")
df_test = pd.read_csv('test.csv', encoding="ISO-8859-1")
df_pro_desc = pd.read_csv('product_descriptions.csv')
```

```python
stemmer = snowballstemmer.EnglishStemmer()
```

```python
def str_stemmer(s):
	return " ".join([stemmer.stemWord(wrd) for wrd in s.lower().split()])
```

```python
def str_common_word(str1, str2):
	return sum(int(str2.find(word)>=0) for word in str1.split())
```

```python
df_all = pd.concat((df_train, df_test), axis=0, ignore_index=True)
```

```python
df_all = pd.merge(df_all, df_pro_desc, how='left', on='product_uid')
```

```python
df_all.head()
```

```python
df_all['search_term'] = df_all['search_term'].map(lambda x:str_stemmer(x))
df_all['product_title'] = df_all['product_title'].map(lambda x:str_stemmer(x))
df_all['product_description'] = df_all['product_description'].map(lambda x:str_stemmer(x))

df_all['len_of_query'] = df_all['search_term'].map(lambda x:len(x.split())).astype(np.int64)

df_all['product_info'] = df_all['search_term']+"\t"+df_all['product_title']+"\t"+df_all['product_description']

df_all['word_in_title'] = df_all['product_info'].map(lambda x:str_common_word(x.split('\t')[0],x.split('\t')[1]))
df_all['word_in_description'] = df_all['product_info'].map(lambda x:str_common_word(x.split('\t')[0],x.split('\t')[2]))

df_all = df_all.drop(['search_term','product_title','product_description','product_info'],axis=1)

```

```python
num_train = df_train.shape[0]
```

```python
df_train = df_all.iloc[:num_train]
df_test = df_all.iloc[num_train:]
id_test = df_test['id']

y_train = df_train['relevance'].values
X_train = df_train.drop(['id','relevance'],axis=1).values
X_test = df_test.drop(['id','relevance'],axis=1).values
```

```python
df_train.drop('relevance',axis=1,inplace=True)
```

```python
y_train
```

```python
import xgboost as xgb
```

```python
xdtest=xgb.DMatrix(df_test,missing=-1)
xdtrain=xgb.DMatrix(df_train,label=y_train,missing=-1)
```

```python
y_train
```

```python
params = {}
params["objective"] ="reg:linear"
params["min_child_weight"] = 120
params["subsample"] = 1
params["colsample_bytree"] = .75
params["silent"] = 0
params["max_depth"] = 4
params["eta"]=0.15
params["gamma"]=1
plst=list(params.items())
```

```python
bst=xgb.train(plst,xdtrain,num_boost_round=400)
```

```python
test_preds=bst.predict(xdtest)
```
